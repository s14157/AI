# 0. Motivation

However, just like numerical methods, AI-generated code comes with trade-offs. The output may be incomplete, subtly incorrect, or logically flawed. Knowing how to guide, evaluate, and revise AI-generated code is an essential modern skill. While it is important to learn how to leverage AI tools to support coding, it is equally, if not more important to critically assess their limitations and drawbacks.

#### Part 1: Handling non-integer input

Recall the Collatz conjecture from Lab 4.

The Collatz conjecture is one of the most famous unsolved problems in mathematics:

> Pick a positive integer `n`, any positive integer. If `n` is even, divide it by 2; if it's odd, multiply it by 3 and add 1. Take that new number and repeat the process, again and again. If you keep repeating this, you'll eventually reach the number 1, regardless of which positive integer `n` is chosen initially. And if you continue the process beyond 1, you will get stuck in a loop. 

$$10 \rightarrow 5 \rightarrow 16 \rightarrow 8 \rightarrow 4 â†’ 2 \rightarrow 1 \rightarrow 4 \rightarrow 2 \rightarrow 1 \rightarrow 4 \rightarrow 2 \rightarrow 1 \rightarrow \dots$$

$$11 \rightarrow 34 \rightarrow 17 \rightarrow 52 \rightarrow 26 \rightarrow 13 \rightarrow 40 \rightarrow 20 \rightarrow 10 \rightarrow 5 \rightarrow 16 \rightarrow 8 \rightarrow 4 \rightarrow 2 \rightarrow 1 \rightarrow 4 \rightarrow \dots $$

The number `n` will travel up and down but eventually end at 1 (at least for all numbers that have ever been tried -- nobody has ever proved that the sequence will terminate). 

The example below is a student submission to Lab 04. Let's run it and see what happens.

```python
def collatz(n): 
    if n != int(n) or n < 0:
        return 'Invalid Input'
    sequence = [n]
    while n != 1: 
        if n % 2 == 0:
            n = n // 2
        else: 
            n = 3 * n + 1
        sequence.append(n)
    return sequence, len(sequence)

challenge_2 = collatz('a')
print(f'Sequence: {challenge_2[0]}')
print(f'Number of steps to get to 1: {challenge_2[1]}')
```

```python
def collatz(n): 
    if n != int(n) or n < 0:
        return 'Invalid Input'
    sequence = [n]
    while n != 1: 
        if n % 2 == 0:
            n = n // 2
        else: 
            n = 3 * n + 1
        sequence.append(n)
    return sequence, len(sequence)

challenge_2 = collatz(0)
print(f'Sequence: {challenge_2[0]}')
print(f'Number of steps to get to 1: {challenge_2[1]}')
```

### Part 1: Cleaning up repetitive code 

Recall plotting the natural modes of vibration of a rectangular membrane of length $a$ and width $b$, fixed along all four edges. Here, we have code below that is repetitive in plotting. Highlight the code below and in the chat (you can use the in-line or the regular chat) ask Copilot to reformat this code and add comments.

```python
def mode_shape(x,y,a,b,m,n):
    return np.sin(m*np.pi*x/a)*np.sin(n*np.pi*y/b)

# length and width 
a = 1
b = 1

# Create x and y 1D grids
x = np.linspace(0,a,1000)
y = np.linspace(0,b,1000)

# Create combined x and y 2D grid
X,Y = np.meshgrid(x,y)

# Create figure of 2x2 3D subplots
fig_7, axs = plt.subplots(2,2,subplot_kw=dict(projection='3d'), figsize=(9,7))

# Plot mode shape on each subplot
Z1 = mode_shape(X,Y,a,b,1,1)
axs[0][0].plot_surface(X,Y,Z1,cmap='coolwarm')
axs[0][0].set_xlabel('X')
axs[0][0].set_ylabel('Y')
axs[0][0].set_zlabel('Z')
axs[0][0].set_title('m = 1, n = 1')

Z2 = mode_shape(X,Y,a,b,1,2)
axs[0][1].plot_surface(X,Y,Z2,cmap='coolwarm')
axs[0][1].set_xlabel('X')
axs[0][1].set_ylabel('Y')
axs[0][1].set_zlabel('Z')
axs[0][1].set_title('m = 1, n = 2')

Z3 = mode_shape(X,Y,a,b,2,1)
axs[1][0].plot_surface(X,Y,Z3,cmap='coolwarm')
axs[1][0].set_xlabel('X')
axs[1][0].set_ylabel('Y')
axs[1][0].set_zlabel('Z')
axs[1][0].set_title('m = 2, n = 1')

Z4 = mode_shape(X,Y,a,b,2,2)
axs[1][1].plot_surface(X,Y,Z4,cmap='coolwarm')
axs[1][1].set_xlabel('X')
axs[1][1].set_ylabel('Y')
axs[1][1].set_zlabel('Z')
axs[1][1].set_title('m = 2, n = 2')

# Set figure title
fig_7.suptitle('Mode shapes', fontsize=16)

plt.show()
```

### Part 2: Finding inefficient steps

From Lab 5, we worked on creating a function to convert binary to decimal numbers. This code snipped below works correctly but has some inefficient or repetitive steps. Ask Copilot to help identify places that can be improved, and provide code suggestions. 

```python
# Inefficient steps for the sign 
def Binary2Num(binary, representation):
    total = 0
    if len(binary) != 8:
        return 'binary should be a string of length 8 made of only 0 and 1'
    for num in binary:
        if num != '0' and num != '1':
            return 'binary should be a string of length 8 made of only 0 and 1'
    if representation == 'UnSign':
        for num in range(8):
            if binary[num] == '1':
                bit = 1
            elif binary[num] == '0':
                bit = 0
            total = total + (bit * (2) ** (7 - num))
        return total
    if representation == 'Sign':
        for num in range(1,8):
            if binary[num] == '1':
                bit = 1
            elif binary[num] == '0':
                bit = 0
            total = total + (bit * (2) ** (7 - num))
        if binary[0] == '1':
            bit = 1
        elif binary[0] == '0':
            bit = 0
        total = total * (-1) ** (bit)
        if binary == '10000000':
            total = -128
            return total
        return total
    if representation == 'TC':
        if binary[0] == '1':
            for num in range(1,8):
                if binary[num] == '1':
                    bit = 1
                elif binary[num] == '0':
                    bit = 0
                total = total + (1 - bit) * (2) ** (7 - num)
            total = (total + 1) * (-1)
        else:
            for num in range(8):
                if binary[num] == '1':
                    bit = 1
                elif binary[num] == '0':
                    bit = 0
                total += (bit * ((2) ** (7-num)))
        return total
    else:
        return 'invalid representation'
    
Binary2Num('11111111', 'TC')
```
